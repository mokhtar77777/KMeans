```python
from kmeans import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
points1 = np.random.multivariate_normal(mean=[0, 0], cov=[[1, 0], [0, 1]], size=100)
points2 = np.random.multivariate_normal(mean=[5, 5], cov=[[1, 0], [0, 1]], size=100)
points3 = np.random.multivariate_normal(mean=[-5, 5], cov=[[1, 0], [0, 1]], size=100)
points4 = np.random.multivariate_normal(mean=[-5, -5], cov=[[1, 0], [0, 1]], size=100)

x = np.vstack((points1, points2, points3, points4))
np.random.shuffle(x)
```

```python
kmeans = KMeans(k=4)
```

```python
clusters = kmeans.fit(x, steps=5, verbose=True)
```

```python
centroids = kmeans.get_centroids()
centroids
```

```python
m = x.shape[0]
colors = ["red", "blue", "green", "black"]
```

```python
for cluster in range(m):
    plt.scatter(x[cluster][0], x[cluster][1], c=colors[clusters[cluster]])
    
plt.scatter(centroids[:, 0], centroids[:, 1], 100, marker='X', linewidth=2, c="yellow")
```

```python
kmeans.predict([[0, 0]])
```
