```python
from kmeans import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(size=(100, 2))
```

```python
kmeans = KMeans(k=4)
```

```python
clusters = kmeans.fit(x, steps=100)
```

```python
kmeans.get_centroids()
```

```python
m = x.shape[0]
colors = ["red", "blue", "green", "black"]
```

```python
for cluster in range(m):
    plt.scatter(x[cluster][0], x[cluster][1], c=colors[clusters[cluster]])
```

```python
kmeans.predict([[0, -3]])
```
